```python
import matplotlib.pyplot as plt
import numpy as np


def visual(nested_arrays):
    plt.matshow(nested_arrays, cmap='viridis')

    for i in range(len(nested_arrays)):
        for j in range(len(nested_arrays[0])):
            plt.text(j, i, str(nested_arrays[i, j]), color='black', ha='center', va='center')

def unhappy_agents(nested_arrays, tol):
    unhappy = []
    for i in range(len(nested_arrays)):
        for j in range(len(nested_arrays[0])):
            if nested_arrays[i, j] != 0:
                rat = np.sum(nested_arrays[i-1:i+2, j-1:j+2] == nested_arrays[i, j]) - 1
                total_agents = np.sum(nested_arrays != 0) - 1  
                if rat / total_agents < tol:
                    unhappy.append((i, j))
    return unhappy

def move_agent(nested_arrays, pos):
    empty_cells = [(i, j) for i in range(len(nested_arrays)) for j in range(len(nested_arrays[0])) if nested_arrays[i, j] == 0]
    new_pos = empty_cells[np.random.randint(len(empty_cells))]
    nested_arrays[new_pos] = nested_arrays[pos]
    nested_arrays[pos] = 0
    return nested_arrays

def step(nested_arrays, tol):
    unhappy = unhappy_agents(nested_arrays, tol)
    if len(unhappy) > 0:
        agent_to_move = unhappy[np.random.randint(len(unhappy))]
        nested_arrays = move_agent(nested_arrays, agent_to_move)
    return nested_arrays


def simulate(rep, tol, N):
    nested_arrays = np.random.randint(0, 3, size=(N, N))
    visual(nested_arrays)
    plt.title('Initial State')
    plt.show()

    for _ in range(rep):
        nested_arrays = step(nested_arrays, tol)
        visual(nested_arrays)
        plt.title('State after Step ' + str(_) + ' of Simulation')
        plt.show()

simulate(10, 0.3, 10)
```
